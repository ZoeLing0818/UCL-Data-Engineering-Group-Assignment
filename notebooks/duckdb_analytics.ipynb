# DuckDB Analytics Notebook

This notebook runs the assignment's DuckDB analytics queries and provides quick visual checks.

## 1. Setup

```python
from pathlib import Path
import duckdb
import pandas as pd

try:
    import seaborn as sns
    import matplotlib.pyplot as plt
    HAS_PLOTS = True
except Exception:
    HAS_PLOTS = False

DB_PATH = Path('../data/analytics.duckdb')
SQL_PATH = Path('../sql/duckdb/queries.sql')

print('DuckDB file exists:', DB_PATH.exists())
print('SQL file exists:', SQL_PATH.exists())
print('Seaborn available:', HAS_PLOTS)
```

## 2. Connect and Load SQL Queries

```python
con = duckdb.connect(str(DB_PATH))
sql_text = SQL_PATH.read_text(encoding='utf-8')
queries = [q.strip() for q in sql_text.split(';') if q.strip()]
print('Total queries:', len(queries))
```

## 3. Run All Queries

```python
results = {}
for i, query in enumerate(queries, start=1):
    df = con.execute(query).fetchdf()
    results[i] = df
    print(f'Query {i}: {len(df)} rows')

list(results.keys())
```

## 4. Preview Results

```python
for i in sorted(results):
    print(f'\n--- Query {i} ---')
    display(results[i].head(20))
```

## 5. Optional Quick Charts

```python
if HAS_PLOTS:
    q2 = results[2].copy()
    plt.figure(figsize=(8, 4))
    sns.barplot(data=q2, x='cvss_severity', y='cve_count', color='#1f6b52')
    plt.title('CVSS Severity Distribution')
    plt.xlabel('CVSS Severity')
    plt.ylabel('CVE Count')
    plt.tight_layout()
    plt.show()

    q3 = results[3].copy()
    plt.figure(figsize=(8, 4))
    sns.barplot(data=q3, x='cvss_severity', y='kev_hit_rate', color='#b56224')
    plt.title('KEV Hit Rate by CVSS Severity')
    plt.xlabel('CVSS Severity')
    plt.ylabel('KEV Hit Rate')
    plt.tight_layout()
    plt.show()
else:
    print('Install seaborn and matplotlib to render charts in this notebook.')
```

## 6. Key Findings Template

- Daily new CVEs show variation over time.
- CVSS severity is concentrated in MEDIUM and HIGH bands.
- KEV hit rate tends to be higher in more severe CVSS categories.
- KEV-flagged CVEs have higher average CVSS and EPSS values than non-KEV CVEs.
